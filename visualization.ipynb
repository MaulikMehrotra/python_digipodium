```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#available styles
print(plt.style.available)
```

```python
#changinh the theme of the plots
plt.style.use('seaborn')
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y)
```

```python
fig=plt.figure(figsize=(6,4))
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y,marker='o',markersize=4, markerfacecolor='blue',markeredgecolor='yellow',markeredgewidth=2)
```

```python
fig=plt.figure(figsize=(6,4))
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y,linestyle=':',color='red',linewidth=1)
```

```python
fig=plt.figure(figsize=(6,4))
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y,linestyle='--',color='blue',linewidth=1,label='sin(x)')
plt.title('timeseries')
plt.xlabel('time')
plt.ylabel('value')
plt.legend()
```
