```python
#removing unwanted columns
cols_to_delete=['_id','pid','url','images']
df.drop(cols_to_delete,axis=1,inplace=True)
#axis=1 vertically
#inplace=True , means update in the memory
```

```python
def clean_price(price):
    if price is None:
        return np.nan
    elif isinstance(price,str):
        price=price.replace(',','')
        if price:
            price=float(price)
        else:
            price=np.nan
        return price
    else:
        return price
    
df['actual_price']=df['actual_price'].apply(clean_price)
```

```python
def clean_discount(discount):
    if discount is None:
        return np.nan
    elif isinstance(discount,str):
        discount=discount.replace('% off','')
        if discount:
            discount=float(discount)
        else:
            discount=np.nan
        return discount
    else:
        return discount
    
df['discount']=df['discount'].apply(clean_discount)
```

```python
def clean_sp(sp):
    if sp is None:
        return np.nan
    elif isinstance(sp,str):
        sp=sp.replace(',','')
        if sp:
            sp=float(sp)
        else:
            sp=np.nan
        return sp
    else:
        return sp

df['selling_price']=df['selling_price'].apply(clean_sp)
```

```python
def clean_rating(rating):
    if rating is None:
        return np.nan
    elif isinstance(rating,str):
        if rating:
            rating=float(rating)
        else:
            rating=np.nan
        return rating
    else:
        return rating
    
df['average_rating']=df['average_rating'].apply(clean_rating)
```

```python

pricesdf['actual_price'].plot.kde()
plt.xlim(0,10000)  #to set the limit of the x-axis
plt.xlabel('Actual Price')
plt.text(4000,0.0006,'Distribution of \nActual Price',fontsize=20,fontweight='bold',color='blue')
conclusion='the distribution is right skewed'
plt.text(4000,0.00055,conclusion,fontsize=10)
plt.show()
```

```python
pricesdf['selling_price'].plot.kde()
plt.xlabel('Selling Price')
plt.xlim(0,12000)
plt.text(4000,0.0012,'Distribution of \nSelling Price',fontsize=20,fontweight='bold',color='Red')
desc='Peak at 0.0016'
plt.text(4000,0.0010,desc,color='blue')
```
