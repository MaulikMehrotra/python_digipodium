```python
def clean_price(price):
    if price is None:
        return np.nan
    elif isinstance(price,str):
        price=price.replace(',','')
        if price:
            price=float(price)
        else:
            price=np.nan
        return price
    else:
        return price
    
df['selling_price']=df['selling_price'].apply(clean_price)
df['actual_price']=df['actual_price'].apply(clean_price)
```

```python
df.groupby('category')['selling_price'].mean()
```
